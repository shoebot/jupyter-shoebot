```python
import shoebot
```

### Before starting

We'll make a convenience function to initialize our Bot object with starting values, so we don't have to repeat them on every cell.

```python
def init_bot():
    bot = shoebot.create_bot()
    bot.size(200,200)
    bot.background(0.2)
    bot.stroke(0.8)
    bot.fill(0.8, 0.3)
    return bot
```

## Rotations

### Around the object's center

```python
bot = init_bot()
bot.transform(bot.CENTER)
for i in range(7):
    bot.rotate(15)
    bot.rect(50, 50, 100, 100)
bot.show()
```

### Around a fixed point

For this, we use the `CORNER` transform mode, which is the default in Shoebot.

Note that this time, we use `translate()` so we can define the point where we want the subsequent shapes to rotate around from.

```python
bot = init_bot()
bot.translate(50, 50)
for i in range(7):
    bot.rotate(15)
    bot.rect(0, 0, 100, 100)
bot.show()
```

## Scaling

### Around the object's center

```python
bot = init_bot()
bot.transform(bot.CENTER)
for i in range(7):
    bot.rect(50, 50, 100, 100)
    bot.scale(.8)
bot.show()
```

### Around a point

```python
bot = init_bot()
bot.translate(50,50)
for i in range(7):
    bot.rect(0, 0, 100, 100)
    bot.scale(.8)
bot.show()
```

## Skewing

### From the object's center

```python
bot = init_bot()
bot.transform(bot.CENTER)
for i in range(7):
    bot.rect(50, 50, 100, 100)
    bot.skew(.2, 0)
bot.show()
```

### From a fixed point

```python
bot = init_bot()
bot.translate(50, 50)
for i in range(7):
    bot.rect(0, 0, 100, 100)
    bot.skew(.2, 0)
bot.show()
```
