# Transform demos

## Rotations

Transforms can work around an object's center by using the `CENTER` transform mode.

```python
size(200,200)
fill('#e55039', 0.2)
transform(CENTER)
for i in range(5):
    rotate(15)
    rect(50, 50, 100, 100)
```

### Around a fixed point

For this, we use the `CORNER` transform mode, which is the default in Shoebot.

Note that this time, we use `translate()` so we can define the point where we want the subsequent shapes to rotate around from.

```python
size(200,200)
fill('#4a69bd', 0.2)
translate(50, 50)
for i in range(7):
    rotate(15)
    rect(0, 0, 100, 100)
```

## Scaling

Around the object center:

```python
size(200,200)
fill('#60a3bc', 0.2)
transform(CENTER)
for i in range(7):
    rect(50, 50, 100, 100)
    scale(.8)
```

And around a fixed point:

```python
size(200,200)
fill('#78e08f', 0.2)
translate(50,50)
for i in range(7):
    rect(0, 0, 100, 100)
    scale(.8)
```

## Skewing

Around center:

```python
size(200,200)
fill('#e58e26', 0.2)
transform(CENTER)
for i in range(7):
    rect(50, 50, 100, 100)
    skew(.2, 0)
```

And around a fixed point:

```python
size(200,200)
fill('#82ccdd', 0.2)
translate(50, 50)
for i in range(7):
    rect(0, 0, 100, 100)
    skew(.2, 0)
```
